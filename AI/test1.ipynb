```python
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False

movies = {
    "流浪地球": [2.01, 4.59, 7.99, 11.83, 16],
    "飞驰人生": [3.19, 5.08, 6.73, 8.10, 9.35],
    "疯狂的外星人": [4.07, 6.92, 9.30, 11.29, 13.03],
    "新喜剧之王": [2.72, 3.79, 4.45, 4.83, 5.11],
    "廉政风云": [0.56, 0.74, 0.83, 0.88, 0.92],
    "神探蒲松龄": [0.66, 0.95, 1.10, 1.17, 1.23],
    "小猪佩奇过大年": [0.58, 0.81, 0.94, 1.01, 1.07],
    "熊出没·原始时代": [1.13, 1.96, 2.73, 3.42, 4.05]
}
plt.figure(figsize=(20, 8))
width = 0.75
bin_width = width / 5

ind = range(0, len(movies))

movie_data = list(movies.values())
print(movie_data)

every_day = []
for i in range(len(movie_data[0])):
    every_day.append([
        movie_data[0][i],
        movie_data[1][i],
        movie_data[2][i],
        movie_data[3][i],
        movie_data[4][i],
        movie_data[5][i],
        movie_data[6][i],
        movie_data[7][i],
    ])

print(every_day)
# 第一种方案

# plt.bar([i - bin_width * 2 for i in ind], every_day[0], width=bin_width, label='第一天')
# plt.bar([i - bin_width for i in ind], every_day[1], width=bin_width, label='第二天')
# plt.bar(ind, every_day[2], width=bin_width, label='第三天')
# plt.bar([i + bin_width for i in ind], every_day[3], width=bin_width, label='第四天')
# plt.bar([i + bin_width * 2 for i in ind], every_day[4], width=bin_width, label='第五天')

# 第二种方案
for index in range(len(every_day)):
    day_tickets = every_day[index]
    xs = [i - (bin_width * (2 - index)) for i in ind]
    plt.bar(xs, day_tickets, width=bin_width, label="第%d 天" % (index + 1))
    # 添加坐标上的数字
    for ticket, x in zip(day_tickets, xs):
        plt.annotate(ticket, xy=(x, ticket), xytext=(x - 0.1, ticket + 0.1))

for index in range(len(every_day)):
    day_tik = every_day[index]
    xs = [i - (bin_width * (2 - index)) for i in ind]
    plt.bar(xs, day_tik, width=bin_width, label="第%d 天" % (index + 1))
    for ticket, x in zip(day_tik, xs):
        plt.annotate(ticket, xy=(x, ticket), xytext=(x - 0.1, ticket + 0.1))
# 设置图例
plt.legend()
plt.ylabel("单位：亿")
plt.title("春节前 5 天电影票房记录")
# 设置 x 轴的坐标
plt.xticks(ind, movies.keys())
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# 创建 DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df = pd.DataFrame(data)

# 查看前两行数据
print(df.head(2))

# 查看 DataFrame 的基本信息
print(df.info())

# 获取描述统计信息
print(df.describe())

# 按年龄排序
df_sorted = df.sort_values(by='Age', ascending=False)
print(df_sorted)

# 选择指定列
print(df[['Name', 'Age']])

# 按索引选择行
print(df.iloc[1:3])  # 选择第二到第三行（按位置）

# 按标签选择行
print(df.loc[1:2])  # 选择第二到第三行（按标签）

# 计算分组统计（按城市分组，计算平均年龄）
print(df.groupby('City')['Age'].mean())

# 处理缺失值（填充缺失值）
df['Age'] = df['Age'].fillna(30)

# 导出为 CSV 文件
df.to_csv('output.csv', index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["axes.unicode_minus"] = False

# 数据
a = np.array([
    [80, 86],  # 第1个学生，平时分80，期末分86
    [82, 80],
    [85, 78],
    [90, 90],
    [86, 82],
    [82, 90],
    [78, 80],
    [92, 94]
])
b = np.array([[0.3], [0.7]])

# 计算最终成绩
final_score = np.dot(a, b)
final_score = final_score.flatten()  # 变成一维

# 划分平时成绩和期末成绩
regular_score = a[:, 0]
final_exam_score = a[:, 1]

x = np.arange(len(a))  # 学生编号

bar_width = 0.25

plt.figure(figsize=(14,6), dpi=100)
plt.bar(x - bar_width, regular_score, width=bar_width, label='平时成绩')
plt.bar(x, final_exam_score, width=bar_width, label='期末成绩')
plt.bar(x + bar_width, final_score, width=bar_width, label='最终成绩')

plt.xlabel('学生编号')
plt.ylabel('成绩')
plt.title('学生成绩对比')
plt.xticks(x, [f'学生{i+1}' for i in x])
plt.legend()
plt.tight_layout()
plt.show()

print('最终成绩为：')
print(final_score)
```

```python
import numpy as np
import matplotlib.pyplot as plt

bar_width = 0.25
x = np.arange(2)     # 两个学生

plt.figure(figsize=(6, 6), dpi=100)
plt.bar(x - bar_width, [80, 82], width=bar_width, label='平时成绩')
plt.bar(x, [86, 80], width=bar_width, label='期末成绩')
plt.bar(x + bar_width, [84.2, 81.4], width=bar_width, label='最终成绩')

# 标记每个柱子底端的x坐标
for i in x:
    plt.text(i - bar_width, 70, f'{i-bar_width:.2f}', ha='center', color='b')
    plt.text(i, 70, f'{i:.2f}', ha='center', color='r')
    plt.text(i + bar_width, 70, f'{i+bar_width:.2f}', ha='center', color='g')

plt.xlabel('学生编号')
plt.ylabel('成绩')
plt.title('演示柱状图错开原理')
plt.xticks(x, [f'学生{i + 1}' for i in x])
plt.legend()
plt.ylim(65, 100)
plt.show()
```
